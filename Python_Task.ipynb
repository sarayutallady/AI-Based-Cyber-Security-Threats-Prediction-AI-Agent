Write a Python program that asks the user to enter a temperature in Celsius and converts it to Fahrenheit and Kelvin.
Then, display all three values neatly using f-strings.

```python
# Taking user input for temperature in Celsius
celsius = float(input("Enter temperature in Celsius: "))

# Conversion formulas
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

# Display results using formatted strings
print(f"Temperature in Celsius: {celsius:.2f}°C")
print(f"Temperature in Fahrenheit: {fahrenheit:.2f}°F")
print(f"Temperature in Kelvin: {kelvin:.2f}K")

```

Write a Python program that stores employee details — name, department, and base salary — in a list of dictionaries.
Then calculate each employee’s final salary after adding a 10% bonus for IT department employees and 5% for others.
Finally, display all employee details neatly in tabular form using

```python
# Prompt 2 — Employee Salary Management System

import pandas as pd

employees = [
    {"Name": "Aarav", "Department": "IT", "Base Salary": 50000},
    {"Name": "Diya", "Department": "HR", "Base Salary": 45000},
    {"Name": "Ravi", "Department": "Finance", "Base Salary": 55000},
    {"Name": "Karan", "Department": "IT", "Base Salary": 60000},
    {"Name": "Sara", "Department": "Marketing", "Base Salary": 48000}
]

for emp in employees:
    if emp["Department"] == "IT":
        emp["Final Salary"] = emp["Base Salary"] * 1.10
    else:
        emp["Final Salary"] = emp["Base Salary"] * 1.05

df = pd.DataFrame(employees)
print(df)

```

Create a Python program that stores monthly sales for a shop in a list.
Find the month with maximum and minimum sales, calculate total and average sales,
and visualize the trend using a line graph.

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [25000, 27000, 30000, 28000, 35000, 33000, 40000, 42000, 39000, 45000, 47000, 50000]

max_sales = max(sales)
min_sales = min(sales)
avg_sales = sum(sales) / len(sales)

print(f"Highest Sales: ₹{max_sales} in {months[sales.index(max_sales)]}")
print(f"Lowest Sales: ₹{min_sales} in {months[sales.index(min_sales)]}")
print(f"Average Monthly Sales: ₹{avg_sales:.2f}")

plt.plot(months, sales, marker='o', color='royalblue')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales (₹)")
plt.grid(True)
plt.show()

```

Plot a line graph of y = x^2 for x values from -10 to 10 using Matplotlib.

```python
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = [i**2 for i in x]

plt.plot(x, y, marker='o', color='teal')
plt.title("Line Graph of y = x²")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

Create a dictionary storing names and marks of 5 students. Print the names of students who scored above 80

Take a string input and count how many vowels and consonants it has.

```python
text = input("Enter a sentence: ").lower()
vowels = "aeiou"
v_count = sum(1 for c in text if c in vowels)
c_count = sum(1 for c in text if c.isalpha() and c not in vowels)

print(f"Vowels: {v_count}")
print(f"Consonants: {c_count}")
```

Write a Python program to print the Fibonacci sequence up to 50

```python
# Prompt 13 — Fibonacci Sequence

a, b = 0, 1
print("Fibonacci sequence up to 50:")
while a <= 50:
    print(a, end=" ")
    a, b = b, a + b
```

Create a NumPy array of 50 random numbers, calculate mean, median, and standard deviation, and visualize the data using a histogram.

```python
# Prompt 14 — NumPy Array Operations

import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(1, 100, 50)
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))

plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Numbers")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```
